## Previo de la Práctica 1

Se le solicita al alumno investigar y colocar la sigueinte información en esta sección:

- Definición de URDF.  
    Un formato de archivo basado en XML utilizado en ROS (Robot Operating System) para describir un robot en términos de su estructura, geometría, articulaciones, sensores, actuadores y propiedades físicas. Permite definir cómo están conectadas las distintas partes del robot y sus movimientos, lo cual facilita la simulación, visualización y control del robot en aplicaciones de robótica.

- Tipos de juntas que están definidas dentro del formato URDF.
    1.- Junta fija (fixed): No permite ningún movimiento entre los dos enlaces conectados. Se utiliza para conexiones rígidas.

    2.-Junta continua (continuous): Permite rotación ilimitada (360 grados) entre los enlaces conectados, como un motor sin límite angular.

    3.-Junta de revoluta (revolute): Permite rotación en un rango limitado (menor a 360 grados) entre los enlaces, como un codo en un brazo robótico.

    4.-Junta prismática (prismatic): Permite movimiento lineal entre los enlaces en una dirección específica, como un pistón.

    5.-Junta flotante (floating): Permite movimientos de traslación y rotación en todas las direcciones; útil para definir la base de un robot en un entorno de simulación.

    6.-Junta planar (planar): Permite traslación en dos ejes y rotación alrededor de un tercer eje perpendicular a los primeros dos.
- Definición de elementos "primitivos" que se pueden utilizar para definir un robot en ROS.
    Son formas geométricas básicas utilizadas para definir las colisiones y las visualizaciones de las partes de un robot en URDF. 

- Que es un nodo y que es un topico en ROS 2
    Nodo: Es una unidad básica de computación en ROS 2 que realiza una función específica dentro de un sistema robótico. Los nodos pueden publicar o suscribirse a tópicos, servicios, acciones, o parámetros, y se utilizan para modularizar el comportamiento de los robots, permitiendo que diferentes funciones (como el control de motores, la captura de sensores, etc.) se manejen de manera separada.

    Tópico: Es un canal de comunicación unidireccional en ROS 2, utilizado para que los nodos intercambien mensajes de forma asíncrona. Los nodos pueden publicar datos a un tópico o suscribirse a él para recibir información. Esta arquitectura es ideal para intercambiar datos de sensores, comandos de control, y otros tipos de mensajes en un sistema robótico.

- Que es RVIZ.
RViz es una herramienta de visualización en ROS que permite representar gráficamente el estado de un robot y su entorno en tiempo real. Se utiliza para ver los modelos de robots, datos de sensores, trayectorias, y otras informaciones críticas durante el desarrollo y prueba de sistemas robóticos. RViz facilita la depuración y comprensión del comportamiento del robot a través de visualizaciones interactivas de elementos como mapas, rutas, transformaciones, y datos de sensores como cámaras y LIDAR.



## Desarrollo de la práctica 
Primero, se creó el workspace donde se desarrollará la práctica. Para ello, se ejecutaron los siguientes comandos en la consola:

```python
cd ROS2Dev
mkdir -p /robot_2025_ws/src
cd robot_2025_ws
colcon build
```

El primer comando nos permite acceder a la carpeta ROS2Dev. A partir de ahí, se crea un directorio llamado robot_2025_ws y, dentro de este, un subdirectorio llamado src. Luego, nos ubicamos en el directorio del workspace y, utilizando colcon build, se compila el workspace.

Después, nos dirigimos a la carpeta src dentro de nuestro workspace con cd src/. Una vez allí, creamos un paquete para ROS 2 con el siguiente comando:

ros2 pkg create --build-type ament_cmake practica1_description

Aquí, ros2 pkg create es el comando para crear paquetes en ROS 2. La opción --build-type ament_cmake especifica que el paquete utilizará el sistema de construcción ament_cmake, el estándar en ROS 2. El último parámetro es el nombre del paquete.

Una vez creado el paquete, debemos crear tres carpetas: urdf, rviz y launch. Para ello, ejecutamos:

mkdir urdf rviz launch

Esto se hace dentro del directorio del paquete. Luego, con el comando code ., abrimos Visual Studio Code en el directorio del paquete para crear dos archivos: uno en la carpeta launch, llamado basic_link_display.launch.xml, y otro en la carpeta urdf, llamado basic_link.urdf.

Código del basic_link_display.launch.xml:

```python
<launch>
    <let name="urdf_path"
        value="$(find-pkg-share practica1_description)/urdf/basic_link.urdf" />
    
    <let name="rviz_config"
        value="$(find-pkg-share practica1_description)/rviz/basic_link_display.rviz" />
    
    <node pkg="robot_state_publisher" exec="robot_state_publisher">
        <param name="robot_description"
            value="$(command 'xacro $(var urdf_path)')" />
    </node>
    <node pkg="joint_state_publisher_gui" exec="joint_state_publisher_gui" />
    
    <node pkg="rviz2" exec="rviz2" output="screen"
        args="-d $(var rviz_config)" />
</launch>
```

Código del basic_link.urdf:

```python
<?xml version="1.0" encoding="utf-8"?>
<robot name="basic_link">
    <link name="base_link">
        <visual>
            <origin xyz="0.0 0.0 0.075" rpy="0.0 0.0 0.0"/>
            <geometry>
                <box size="0.1 0.12 0.15"/>             
            </geometry>
            <material name="brown">
                <color rgba="0.1 0.43 0.43 1.0" />
            </material>      
        </visual>
        <collision>
            <origin xyz="0.0 0.0 0.075" rpy="0.0 0.0 0.0"/>
            <geometry>
                <box size="0.1 0.12 0.15"/>             
            </geometry>          
        </collision>
        <inertial>
            <mass value="0.5"/>
            <origin xyz="0.0 0.0 0.0" rpy="0.0 0.0 0.0"/>
            <inertia ixx="0.1" ixy="0.0" ixz="0.0" iyy="1.0" iyz="0.0" izz="1.0"/> 
        </inertial>
    </link>

    <link name="link_1">
        <visual>
            <origin xyz="0.025 0.0 0.25" rpy="0.0 0.0 0.0"/>
            <geometry>
                <box size="0.05 0.05 0.5"/>             
            </geometry>
            <material name="red">
                <color rgba="0.5 0.0 0.0 1.0" />
            </material>      
        </visual>
        <collision>
            <origin xyz="0.025 0.0 0.25" rpy="0.0 0.0 0.0"/>
            <geometry>
                <box size="0.05 0.05 0.5"/>             
            </geometry>          
        </collision>
        <inertial>
            <mass value="0.3"/>
            <origin xyz="0.0 0.0 0.0" rpy="0.0 0.0 0.0"/>
            <inertia ixx="0.1" ixy="0.0" ixz="0.0" iyy="1.0" iyz="0.0" izz="1.0"/> 
        </inertial>
    </link>

    <joint name="base_link_to_link_1" type="revolute">
        <parent link="base_link"/>
        <child link="link_1"/>
        <axis xyz="1 0 0"/>
        <limit lower="-3.14159" upper="3.14159" velocity="50.0" effort="1000.0"/>
        <origin xyz="0.05 0 0.15" rpy="0 0 0"/>
    </joint>
</robot>
```

Una vez terminado de programar, regresamos al directorio del workspace y descargamos dos plugins necesarios ejecutando:

sudo apt install ros-humble-joint-state-publisher-gui
sudo apt install ros-humble-xacro
sudo apt-get update

Luego, compilamos nuevamente el workspace con:

colcon build
source install/setup.bash

Si se siguieron los pasos correctamente, no deberían existir errores al ejecutar el URDF en RViz. Para visualizarlo, ejecutamos:

ros2 launch practica1_description basic_link_display.launch.xml

Esperamos unos segundos para que RViz se abra. Configuramos el "Fixed Frame" seleccionando base_link. Agregamos el URDF con el botón Add, buscamos "RobotModel" y seleccionamos. En la configuración de "RobotModel", cambiamos DescriptionSource a file y seleccionamos la ubicación de nuestro archivo URDF.

Como parte de la práctica, se debe agregar otro eslabón más al robot, ya que actualmente cuenta solo con una junta y dos eslabones. Para hacerlo, editamos el archivo basic_link.urdf y agregamos el eslabón link_2 con su masa, propiedades inerciales y visuales, además de una junta adicional que conecta link_1 con link_2.
